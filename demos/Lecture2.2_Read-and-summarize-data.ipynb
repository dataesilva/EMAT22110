### Step 1: Load in required packages

```python
import json
import statistics

with open(r"D:\EMAT_22110\FB Data Silva\facebook-davidsilva100046\ads_and_businesses\advertisers_who_uploaded_a_contact_list_with_your_information.json") as a:
    dat = json.load(a)
print(dat)
```

To display the Keys included in the dictionary named dat we use the `.key()` method

```python
dat.keys()
```

To determine the type of data stored in the 'custom_audiences' key, use the `type()` function. The output is `list` which means a collection of data indexed starting at 0.

```python
type(dat['custom_audiences'])
```

To find out what type of data is stored in each element in the list, we can use `type()` again, but select the 0th element from the list using another set of brackets. The output is `str` which means string.

```python
type(dat['custom_audiences'][0])
```

We can also determine how many elements are included in this list using the `len()` function.

```python
len(dat['custom_audiences'])
```

In order to find out the length of each string in our list, we need to loop across the entire list and look at each element individually. To do this, we will use the [list comprehension approach](https://docs.python.org/2/tutorial/datastructures.html#list-comprehensions) shown below. This applies `len()` to each element and stores the output for all elements as another list that I've decided to name `namlens`.  

```python
namelens = [len(x) for x in dat['custom_audiences']]
namelens
```

We can now calculate some statistics you may be familiar with.

```python
max(namelens)
```

```python
statistics.mean(namelens)
```
