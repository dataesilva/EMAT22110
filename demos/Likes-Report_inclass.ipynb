## With either data structure, you will be able to follow along exactly from here with groupby() and count()

```python
count_elikes = elikes_df.groupby('accounts').count()
count_elikes
```

```python
count_elikes.sort_values('timestamp' ,ascending=False)
```

## Remember to import matplotlib and seaborn for the three plots. The bar and pie plot are shown below. The displot was covered in prior classes.

```python
from matplotlib import pyplot as plt
```

```python
sortcount = count_elikes.sort_values('timestamp' ,ascending=False)
labels = sortcount['timestamp'].keys()
vals = sortcount['timestamp']
width = .75

fig, ax = plt.subplots(figsize=(10, 10), dpi= 80, facecolor='w', edgecolor='k')
ax.bar(labels, vals, width)
ax.set_ylabel('Likes')
plt.xticks(rotation=90, fontsize = 'small')
plt.show()
```

```python
fig2, ax2 = plt.subplots()
ax2.pie(vals, labels = labels)
ax2.axis('equal')
plt.show()
```

## If you want to make your plot more readable, you can use this cell below. You may need to adjust from 50 in the first line up or down depending on your total number of likes.

```python
toplikes = count_elikes['timestamp'] > 50
countstop, countsbot = sortcount[toplikes], sortcount[-toplikes]
other = countsbot.sum()

piedat = countstop.append(pd.DataFrame(other['timestamp'], columns = ['timestamp'], index = ['other < 50']))
piedat
```

```python
vals = piedat['timestamp']
labels = piedat['timestamp'].keys()


fig3, ax3 = plt.subplots()
ax3.pie(vals, labels = labels)
ax3.axis('equal')
plt.show()
```
